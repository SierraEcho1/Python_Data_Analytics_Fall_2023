```python
#Removing missing values from data set
missing_values=["na","n/a","--"]
df=pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv', na_values=missing_values)
```

```python
#Output of removed missing values. Do not see a change in data set
df
```

```python
#Drop column with missing values
drop_cols = ['weight']
```

```python
df1=df.drop(drop_cols, axis=1)
```

```python
#Output of data with removed "Weight" column
df1
```

```python
#Race taking the most medication
sns.jointplot(x='race', y='num_medications', data=df1)
```

```python
#Create data subset 
df2=df[['race', 'gender','age','num_medications']]
```

```python
#Output of new data set 
df2
```

```python
df2=df2.rename(columns={'race': 'Race','gender': 'Gender','age': 'Age','num_medications': 'Number of Rx'})
```

```python
#Rename of columns
df2
```

```python
#Average number of Rx taken by patient
df2["Number of Rx"].mean()
```

```python
#Comparison point - Number of Rx by Race
df2.groupby(['Race'])['Number of Rx'].count().sort_values(ascending=False)
```

```python
#Graph of comparison point

df2.groupby(['Race'])['Number of Rx'].count().sort_values(ascending=False).plot.bar()
```

```python
#Shows most patients take b/w 10-20 Rx
sns.displot(df2['Number of Rx'])
```
